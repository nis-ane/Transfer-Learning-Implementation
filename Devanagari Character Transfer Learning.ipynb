# Transfer Learning for Recognition of Devanagari Characters using VGG pretrained Model

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
import keras
import keras_utils
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout, InputLayer
from tensorflow.keras import optimizers
import pandas as pd
from random import shuffle
from tqdm import tqdm
import os
import cv2
from keras.applications import vgg16
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_gen = ImageDataGenerator(rescale = 1/255)
train_img_gen = img_gen.flow_from_directory('DevanagariHandwrittenCharacterDataset/Train/', target_size = (32,32), batch_size =16, class_mode='categorical', color_mode ='rgb')
validate_img_gen = img_gen.flow_from_directory('DevanagariHandwrittenCharacterDataset/Validate/', target_size = (32,32), batch_size =16, class_mode='categorical', color_mode ='rgb')
```

```python
train_img_gen.class_indices
```

```python
test_data = 'DevanagariHandwrittenCharacterDataset/Test/'
```

```python
def test_pictures(input_size, channel):
    predict_images = []
    for i in tqdm(os.listdir(test_data)):
        path = os.path.join(test_data, i)#path to the test image in folder
        if channel==1:
            scale = cv2.IMREAD_GRAYSCALE
        else:
            scale = cv2.IMREAD_COLOR
                
        img = cv2.imread(path, scale)
        img = cv2.resize(img, dsize= (input_size,input_size))
        predict_images.append([np.array(img)])   #appended([image1],[image2],.....)   
    return predict_images
```

```python
def test_model(model, input_size, channel):
    testing_images = test_pictures(input_size, channel)
    test_img_data = np.array([i[0] for i in testing_images]).reshape(-1,input_size, input_size,channel)
    label=[]
    for images in test_img_data:
        images = np.expand_dims(images, axis=0)/255
        y_pred = model.predict(images)
        label.append(np.argmax(y_pred, axis=1))

    count=0
    for i in testing_images:
        i= np.reshape(i,(input_size, input_size, channel))
        plt.imshow(np.squeeze(i))
        for key,values in train_img_gen.class_indices.items():
            if label[count] == values:
                k = key
        plt.xlabel(k)
        count=count+1
        plt.show()
```

```python
def plot_history(history):
    fig, (ax1,ax2)= plt.subplots(1,2,figsize=(15,7))
    ax1.plot(history.history['loss'])
    ax1.plot(history.history['val_loss'])
    ax1.set_title('Model Loss')
    ax1.set(xlabel='epoch', ylabel='loss')
    ax1.legend(['train', 'val'], loc='upper left')
    ax2.plot(history.history['acc'])
    ax2.plot(history.history['val_acc'])
    ax2.set_title('Model Accuracy')
    ax2.set(xlabel='epoch', ylabel='accuracy')
    ax2.legend(['train', 'val'], loc='upper left')
```

```python
from keras.models import Model
```

```python
input_shape = (32,32,3)
```

```python
vgg = vgg16.VGG16(include_top=False, weights='imagenet', input_shape=input_shape)
```

```python
output = vgg.layers[-1].output
output = keras.layers.Flatten()(output)
vgg_model = Model(vgg.input, output)
```

```python
vgg_model.trainable = False
```

```python
for layer in vgg_model.layers:
    layer.trainable = False
```

```python
import pandas as pd
```

```python
pd.set_option('max_colwidth', -1)
```

```python
layers = [(layer, layer.name, layer.trainable) for layer in vgg_model.layers]
```

```python
pd.DataFrame(layers, columns=['Layer Type', 'Layer Name', 'Layer Trainable'])
```

```python
vgg_model.summary()
```

```python
input_shape = vgg_model.output_shape[1]
```

```python
def make_model():
    new_output = keras.layers.Dense(512, activation='relu')(vgg_model.output)
    new_output = keras.layers.Dense(46, activation='softmax')(new_output)
    model = keras.engine.training.Model(vgg_model.inputs, new_output)
    return model
```

```python
model = make_model()
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics =['acc'])
```

```python
history = model.fit_generator(train_img_gen, steps_per_epoch =3000, epochs = 10, validation_data = validate_img_gen, validation_steps=20, verbose=2)
```

```python
plot_history(history)
```

```python
test_model(model, 32,3)
```

```python
model_json = model.to_json()

with open("VGG total frozen model.json", "w")as json_file:  #model is saved
    json_file.write(model_json)
    
model.save_weights("weights for Devanagri classification with total frozen VGG layers.h5") #weights are saved
```
